La energía es un recurso esencial para el funcionamiento de la sociedad moderna. La producción y el consumo de electricidad influyen en diversos sectores, desde la industria hasta el uso doméstico. En un contexto donde la demanda energética sigue en aumento, es importante analizar cómo se genera la energía y qué factores afectan su eficiencia y sostenibilidad.

La *Global Power Plant Database* es una fuente abierta de información que recopila datos sobre más de 30,000 plantas de energía en 164 países. Contiene información detallada sobre diversos tipos de plantas, desde las tradicionales (carbón, gas, nuclear) hasta las renovables (solar, eólica, hidroeléctrica). Cada planta está geolocalizada e incluye datos sobre su capacidad de generación, tipo de combustible, propietario y detalles operativos.

Esta base de datos es una herramienta poderosa para distintos tipos de análisis, como los siguientes:

- **Distribución Geográfica de la Energía:** Permite mapear la ubicación de las plantas para entender mejor la infraestructura energética en diferentes regiones e identificar áreas con menor acceso a fuentes de energía.
- **Impacto Ambiental:** Facilita el estudio de emisiones de plantas térmicas y su efecto en el medioambiente, permitiendo la integración con datos ecológicos y demográficos.
- **Eficiencia y Mercado Energético:** Ayuda a evaluar el desempeño de las plantas, analizar la concentración de la propiedad y comparar modelos de gestión estatal y privada.
- **Efecto de Políticas y Regulaciones:** Permite estudiar el impacto de normativas gubernamentales y acuerdos internacionales en la generación de energía, especialmente en la transición hacia fuentes renovables.
- **Capacidad de generación:** Comparación de la producción energética entre diferentes regiones y tipos de plantas.  
- **Distribución de fuentes de energía:** Evaluación del uso de combustibles fósiles frente a las energías renovables.  
- **Eficiencia operativa:** Detección de plantas subutilizadas y oportunidades de optimización.  
- **Tendencias y predicciones:** Análisis de patrones históricos para estimar necesidades futuras.  

Este estudio busca comprender mejor la distribución y eficiencia de la generación energética en el mundo, proporcionando una base para futuras investigaciones sobre sostenibilidad y transición energética.

#### **Desafíos en la Generación Energética**  

La generación de energía enfrenta varios desafíos relacionados con la eficiencia, la sostenibilidad y la equidad en la distribución. A pesar del crecimiento de las energías renovables, muchas regiones aún dependen de combustibles fósiles, lo que plantea problemas tanto ambientales como económicos.

- **Distribución desigual:** Mientras algunas zonas cuentan con un suministro energético estable, otras enfrentan limitaciones de acceso.  
- **Variabilidad en la generación:** Factores climáticos y tecnológicos pueden afectar la estabilidad del suministro eléctrico.  

A partir de estos problemas, surgen diversas oportunidades para mejorar la generación energética:

- **Optimización del uso de las plantas:** Identificación de áreas donde se puede mejorar la eficiencia operativa.  
- **Fomento de energías renovables:** Evaluación de la viabilidad de reemplazar fuentes fósiles por alternativas más sostenibles.  
- **Planificación energética:** Uso de modelos predictivos para anticipar fluctuaciones en la producción y mejorar la estabilidad del sistema.

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Asegurar que latitude y longitude no tengan valores nulos
data = data.dropna(subset=['latitude', 'longitude'])

# Convertir 'capacity_mw' a numérico y eliminar valores NaN
data['capacity_mw'] = pd.to_numeric(data['capacity_mw'], errors='coerce')
data = data.dropna(subset=['capacity_mw'])

# Crear geometría a partir de las coordenadas
geometry = [Point(xy) for xy in zip(data['longitude'], data['latitude'])]
geo_df = gpd.GeoDataFrame(data, geometry=geometry)

# Cargar el mapa mundial
world = gpd.read_file("https://naturalearth.s3.amazonaws.com/110m_cultural/ne_110m_admin_0_countries.zip")

# Ajustar el tamaño de los puntos de forma proporcional a capacity_mw
geo_df['size'] = geo_df['capacity_mw'] / geo_df['capacity_mw'].max() * 50  # Normalizar tamaño

# Graficar
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)  # Dibujar límites de países

# Dibujar puntos con colores diferenciados y tamaños proporcionales
geo_df.plot(column='capacity_mw', ax=ax, markersize=geo_df['size'], legend=True, cmap='plasma', alpha=0.8, edgecolor="green")

# Etiquetas
plt.title('Ubicación de las Plantas de Energía y su Capacidad (MW)')
plt.xlabel('Longitud')
plt.ylabel('Latitud')

plt.show()

```

```python
# Tipos de combustible
tipos_combustible = data['primary_fuel'].value_counts()
plt.figure(figsize=(10, 6))
sns.countplot(y='primary_fuel', data=data, order=tipos_combustible.index , palette = "magma")
plt.title('Distribución Plantas')
plt.xlabel('Cantidad de Plantas')
plt.ylabel('Tipo de Combustible')
plt.show()
```

```python
# Define las columnas que representan los datos de generación anuales
generation_columns = ['generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015',
                      'generation_gwh_2016', 'generation_gwh_2017', 'generation_gwh_2018', 'generation_gwh_2019']

# Agrega la producción total de energía por año
yearly_production = data[generation_columns].sum()

# Convertir el índice en años enteros para la visualización
yearly_production.index = [int(year.split('_')[-1]) for year in yearly_production.index]

# Crear un gráfico de línea para los datos de producción energética anuales
plt.figure(figsize=(12, 5))
sns.lineplot(x=yearly_production.index, y=yearly_production.values, marker='o', color='b')
plt.title('Total Energy Production by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Total Production (GWh)', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=1)
plt.xticks(yearly_production.index)  # Asegúrate de que solo se usen años enteros como marcas en el eje x
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar datos por país y tipo de combustible
grouped_data = data.groupby(['country_long', 'primary_fuel'])[['capacity_mw', 'generation_gwh_2019']].sum().reset_index()

# Calcular eficiencia de producción (GWh por MW instalado)
grouped_data['efficiency'] = grouped_data['generation_gwh_2019'] / grouped_data['capacity_mw']
grouped_data
```

```python
# Identificar países y tipos de combustible con eficiencia baja
low_efficiency = grouped_data[grouped_data['efficiency'] < grouped_data['efficiency'].quantile(0.25)]

print("Países y tipos de combustible con baja eficiencia:")
low_efficiency
```

```python
# Comparar eficiencia entre fuentes renovables y no renovables
renewables = ['Hydro', 'Solar', 'Wind', 'Biomass', 'Geothermal']
data['is_renewable'] = data['primary_fuel'].isin(renewables)

# Calcular eficiencia promedio para renovables y no renovables
efficiency_comparison = data.groupby('is_renewable')[['capacity_mw', 'generation_gwh_2019']].sum()
efficiency_comparison['efficiency'] = efficiency_comparison['generation_gwh_2019'] / efficiency_comparison['capacity_mw']

print("Comparación de eficiencia entre fuentes renovables y no renovables:")
print(efficiency_comparison)
```

### Análisis Detallado

#### Capacidad Instalada

- **Fuentes No Renovables**: La capacidad instalada para las fuentes no renovables es significativamente mayor que la de las fuentes renovables. Esto incluye plantas que utilizan combustibles fósiles como carbón, petróleo y gas natural, así como plantas nucleares.
- **Fuentes Renovables**: Las fuentes renovables tienen una menor capacidad instalada en comparación, lo que incluye tecnologías como solar, eólica, hidroeléctrica, geotermia, biomasa y residuos.

#### Generación Anual de Energía

- **Fuentes No Renovables**: Las fuentes no renovables generan una cantidad sustancialmente mayor de energía en comparación con las renovables, lo que se espera debido a su mayor capacidad instalada y, a menudo, su capacidad para operar a mayor capacidad continua.
- **Fuentes Renovables**: Aunque las fuentes renovables tienen una menor generación total, esta es una contribución significativa considerando la menor capacidad instalada.

#### Eficiencia

La eficiencia se define como la relación entre la generación anual de energía y la capacidad instalada, y se calcula como:

$\text{Eficiencia} = \frac{\text{Generación Anual de Energía (GWh)}}{\text{Capacidad Instalada (MW)}}$

- **Fuentes No Renovables**: Una eficiencia de 0.819210 indica que, en promedio, las plantas no renovables generan 819.210 GWh por cada 1,000 MW de capacidad instalada. Esto refleja una alta eficiencia operativa, ya que las plantas no renovables, especialmente las nucleares, suelen funcionar a una capacidad constante y alta.
- **Fuentes Renovables**: Una eficiencia de 0.445026 indica que las plantas renovables generan 445.026 GWh por cada 1,000 MW de capacidad instalada. Esta menor eficiencia puede ser atribuida a la variabilidad inherente en la disponibilidad de recursos renovables, como el sol y el viento, que no siempre están disponibles.

### Interpretación y Aplicación

#### Comparación de Eficiencia

1. **Fuentes No Renovables**:
   - **Mayor Eficiencia**: Las fuentes no renovables tienden a ser más eficientes en términos de utilización de capacidad instalada debido a la naturaleza constante y predecible de su producción.
   - **Operación Continua**: Plantas de carbón, gas, y nuclear pueden operar continuamente, lo que les permite mantener una alta eficiencia.

2. **Fuentes Renovables**:
   - **Menor Eficiencia**: Las fuentes renovables tienen una menor eficiencia debido a la variabilidad en la disponibilidad de recursos naturales. Por ejemplo, la energía solar solo se genera cuando hay luz solar y la energía eólica depende de la presencia de viento.
   - **Dependencia de Recursos Naturales**: La eficiencia de las plantas renovables está sujeta a condiciones climáticas y geográficas, lo que puede limitar su capacidad para operar a plena capacidad todo el tiempo.

#### Implicaciones para la Planificación y la Política Energética

1. **Optimización de la Capacidad**:
   - **Plantas No Renovables**: Continuar optimizando la operación de plantas no renovables puede ayudar a mantener una alta eficiencia. Sin embargo, es crucial considerar el impacto ambiental y las emisiones de carbono asociadas.
   - **Plantas Renovables**: Incrementar la capacidad instalada de plantas renovables y mejorar las tecnologías de almacenamiento puede ayudar a aumentar su eficiencia. Almacenamiento de energía puede permitir una utilización más constante de la energía generada.

2. **Transición Energética**:
   - **Inversiones en Renovables**: Aumentar las inversiones en energías renovables es clave para reducir las emisiones de carbono y mitigar el cambio climático, aunque esto puede implicar aceptar una menor eficiencia operativa inicial.
   - **Tecnologías de Mejora**: Invertir en tecnologías que mejoren la predictibilidad y el almacenamiento de energías renovables puede ayudar a cerrar la brecha de eficiencia.

3. **Políticas de Apoyo**:
   - **Subsidios y Incentivos**: Ofrecer subsidios e incentivos para la instalación de nuevas capacidades de energías renovables y para la investigación en tecnologías de almacenamiento puede mejorar la adopción y la eficiencia de las energías renovables.
   - **Regulación de Emisiones**: Implementar regulaciones más estrictas sobre las emisiones de plantas no renovables puede incentivar una mayor inversión en renovables y en tecnologías más limpias.

### Conclusión

La comparación de eficiencia entre fuentes renovables y no renovables revela que, aunque las plantas no renovables son actualmente más eficientes en términos de utilización de capacidad, las plantas renovables representan una oportunidad crítica para una transición hacia un sistema energético más sostenible. Mejorar la eficiencia de las plantas renovables a través de inversiones en tecnología y almacenamiento, junto con políticas de apoyo, puede ayudar a aumentar su participación en el mix energético global y reducir la dependencia de combustibles fósiles.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calcular_estadisticas(df):
    """
    Calcula la matriz de correlación, varianza y media de las columnas numéricas de un DataFrame.
    """
    # Seleccionar solo columnas numéricas
    df_numerico = df.select_dtypes(include=[np.number])

    # Calcular matriz de correlación
    matriz_correlacion = df_numerico.corr()

    # Calcular varianza
    varianza = df_numerico.var()

    # Calcular media
    media = df_numerico.mean()

    return matriz_correlacion, varianza, media


def visualizar_correlacion(matriz_correlacion):
    """
    Visualiza la matriz de correlación usando un heatmap.
    """
    plt.figure(figsize=(12, 8))
    sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
    plt.title('Matriz de Correlación')
    plt.show()


df = data
matriz_correlacion, varianza, media = calcular_estadisticas(df)
visualizar_correlacion(matriz_correlacion)

#Imprimir resultados
print("Varianza:\n", varianza)
print("Media:\n", media)
```
